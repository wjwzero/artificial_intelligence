## Action3 使用WordCloud对新闻进行词云分析

```python
from wordcloud import WordCloud
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import jieba
```

```python
# 去掉信息量低的词
def remove_low_info_words(f):
    low_info_words = ['之一', '表明', '表示', '没有', '状态', '他们', '可能', '这些', '这个', '来自', '如果', '造成', '拥有', '越来越', '最高']
    for low_info_word in low_info_words:
        f = f.replace(low_info_word, '')
    return f
```

```python
# 生成词云
def create_word_cloud(f):
    print('根据词频，开始生成词云！')
    f = remove_low_info_words(f)
    cut_text = jieba.cut(f)
    cut_text = ' '.join(cut_text)
    wc = WordCloud(
        font_path = "simhei.ttf",
        max_words = 100,
        width = 3000,
        height = 2200,
    )
    wordcloud = wc.generate(cut_text)
    # 写词云图片
    wordcloud.to_file("worldcloud.jpg")
    # 显示词云文件
    plt.imshow(wordcloud)
```

```python
# 数据加载
with open('news.txt', encoding = 'utf-8', mode = 'r') as file:
    data = file.read().replace('\n', '')
# 生成词云
create_word_cloud(data)
```
